```python

def get_overview():
    response = httpx.get(url)
    return {"total_games_in_progress": response.json()["totalGamesInProgress"],
            "total_games": response.json()["totalGames"]}#%%


get_overview()
```

```python
def get_days_games():
    response = httpx.get(url)
    games = response.json()["dates"][0]['games']
    return games

games = get_days_games()
```
